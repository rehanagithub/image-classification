```python
import os
import zipfile
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# Define the path to your zip file
zip_file_path = 'burger.zip'

# Define the folder where you want to extract the images
extracted_folder = 'extracted_images'

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)

# Function to preprocess images
def preprocess_image(image_path, image_size):
    image = cv2.imread(image_path)
    image = cv2.resize(image, image_size)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = image.astype('float32') / 255.0  # Normalize pixel values
    return image

# Prepare data and labels
data = []
labels = []
label_encoder = LabelEncoder()

for folder_name in os.listdir(extracted_folder):
    folder_path = os.path.join(extracted_folder, folder_name)
    if os.path.isdir(folder_path):
        for image_file in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_file)
            # Preprocess image
            image = preprocess_image(image_path, (224, 224))  # MobileNetV2 input size
            data.append(image)
            labels.append(folder_name)

# Convert data and labels to numpy arrays
data = np.array(data)
labels = np.array(labels)

# Encode labels into numerical format
labels = label_encoder.fit_transform(labels)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Load MobileNetV2 base model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add custom classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
predictions = Dense(len(np.unique(labels)), activation='softmax')(x)

# Combine base model and custom head
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
# Evaluate the model on the test dataset
loss, accuracy = model.evaluate(X_test, y_test)

# Print the test loss and accuracy
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')

# If you want additional metrics, you can compute them manually using sklearn.metrics
from sklearn.metrics import classification_report, confusion_matrix

# Get the model predictions on the test data
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Print classification report
print(classification_report(y_test, y_pred_classes))

# Print confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
import matplotlib.pyplot as plt

# Define class names
class_names = label_encoder.classes_

# Function to plot images with labels
def plot_images(images, true_labels, pred_labels):
    plt.figure(figsize=(10, 10))
    for i in range(min(25, len(images))):
        plt.subplot(5, 5, i + 1)
        plt.imshow(images[i])
        plt.title(f'True: {class_names[true_labels[i]]}\nPredicted: {class_names[pred_labels[i]]}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Get model predictions on the test data
predictions = model.predict(X_test)
predicted_labels = np.argmax(predictions, axis=1)

# Plot some sample images with true and predicted labels
plot_images(X_test, y_test, predicted_labels)

# Find misclassified images
misclassified_images = X_test[y_test != predicted_labels]
true_labels_misclassified = y_test[y_test != predicted_labels]
predicted_labels_misclassified = predicted_labels[y_test != predicted_labels]

# Plot misclassified images
plot_images(misclassified_images, true_labels_misclassified, predicted_labels_misclassified)

```
